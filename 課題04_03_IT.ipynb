# **TRPG風ゲーム**

## **概要**

TRPG風のゲームです。

ステータスが割り振られており、それによって行動が分岐するようになっています。


## **操作方法**

input()を多々使います。ひらがなで入力を求められているときはときはひらがなで、前に数字が割り振られているときは数字で選択してください。数字の選択では半角しか対応していません。対応していない入力の場合もう一度質問を繰り返します。

「戦闘」のステータスが低い場合、運がよくないとエンド1には行けなくなっています。
エンド1に行きたい場合は何度かリセマラすることを推奨します。


## **制作者のコメント**

●注力したこと

　・自分の選択で自由に部屋を移動できること

　　　最初は順番に部屋を回るだけにしようと思っていました。



　・ステ振り

　　　基礎ステータスの振り方は、TRPGのステ振りと同じになっています。

私はTRPGが好きなのですが、好きなだけでやったことのないにわかだったので、ちょくちょく調べながらつくりました。自分には技術もないので時間がかかってしましました。

選択で自由に行動できるようにできたので、ゲーム性を高められたと思います。

時間が足りず省いてしまったものが多々あるので悔しいですが、今は完成したことに満足しています。次は「発狂」や技能を自分で振れるよにできたらいいなと思います。


●反省点

　・一度調べた部屋はもう一度調べられないようにしたのですが、部屋の中の調べる選択肢は何度でも調べられるようになってしまっているので、そこを修正したかったです。

　・修羅との戦闘を時間がなくて省いてしまいました。

　・はじめ、武器庫に行ったとき「刀」が入手でき、それによってダメージボーナス出すつもりだったのですが、時間がなく実装できず、「刀」は消すことにしました。

```python
%reset -f
import random

def roll_dice(num_dice, num_sides):
    """サイコロを振る関数"""
    return sum(random.randint(1, num_sides) for _ in range(num_dice))

class Character:
    def __init__(self, name):
        self.name = name
        self.STR = roll_dice(3, 6)  # 筋力
        self.CON = roll_dice(3, 6)  # 体力
        self.POW = roll_dice(3, 6)  # 精神力
        self.DEX = roll_dice(3, 6)  # 器用さ
        self.APP = roll_dice(3, 6)  # 外見
        self.SIZ = roll_dice(2, 6) + 6  # 体格
        self.INT = roll_dice(2, 6) + 6  # 知性
        self.EDU = roll_dice(3, 6) + 3  # 教育
        self.luck = self.POW * 5  # 幸運
        self.sanity = self.POW * 5  # 正気度
        self.idea = self.INT * 5  # アイデア
        self.knowledge = self.EDU * 5  # 知識
        self.durability = (self.CON + self.SIZ)  # HP（耐久力）
        self.kikimimi = self.CON * 5  # 聞き耳
        self.mebosi = self.INT * 5  # 目星
        self.sentou = self.STR * 5  # 戦闘
        self.kaihi = self.DEX * 5  # 回避
        self.star = 0  # ★
        self.inventory = []  # アイテム欄

    def display_status(self):
        print(f"名前: {self.name}")
        print(f"筋力: {self.STR}, 体力: {self.CON}, 精神力: {self.POW}, 器用さ: {self.DEX}, 外見: {self.APP}, 体格: {self.SIZ}, 知性: {self.INT}, 教育: {self.EDU}, 幸運: {self.luck}, 正気度: {self.sanity}, アイデア: {self.idea}, 知識: {self.knowledge}, 耐久力: {self.durability}, 聞き耳: {self.kikimimi}, 目星: {self.mebosi}, 回避: {self.kaihi}, 戦闘: {self.sentou}")
        print("アイテム: ")

    def add_item(self, item_name):
        """アイテムをアイテム欄に追加する"""
        self.inventory.append(item_name)
        print(f"{item_name}をアイテム欄に追加しました。")

    def display_inventory(self):
        """アイテム欄を表示する"""
        if self.inventory:
            print("アイテム欄:")
            for item in self.inventory:
                print(f"- {item}")
        else:
            print("アイテム欄は空です。")

class Room:
    def __init__(self, name, description):
        self.name = name
        self.description = description
        self.item_searched = False  # アイテムを調べたかどうか

    def explore(self):
        print(self.description)

    def actions(self):
        """部屋内でのアクションを表示する"""
        print("\n何をしますか？")
        if self.name == "和室":
            print("1: 机を調べる")
            print("2: 本棚を調べる")
            print("3: 部屋を出る（右の部屋、左の部屋、奥の部屋へ移動）")
        elif self.name == "右の部屋":
            print("1: 本棚を調べる")
            print("3: 部屋を出る（和室へ戻る）")
        elif self.name == "左の部屋":
            print("1: 武器")
            print("3: 部屋を出る（和室へ戻る）")
        elif self.name == "奥の部屋":
            print("1: 戦闘する")
            print("2: 逃げる")

def text_adventure_game():
    # キャラクターの名前を入力
    character_name = input("キャラクターの名前を入力してください: ")
    player = Character(character_name)
    player.display_status()

    print(" ")
    print("貴方は長引いた仕事をやっと終わらせ帰宅中のそんなある日。")
    print(" ")

    # 初期の和室の作成
    main_room = Room("和室", "一般的なよくある和室だ。机と本棚があり、部屋の右、左、奥にそれぞれ扉がある。")
    right_room = Room("右の部屋", "【図書室】右の扉を開けて見るとそこは図書室だった。いくつかの本棚がある、探してみればなにか情報が得られるかもしれない")
    left_room = Room("左の部屋", "【武器庫】左の扉を開けるとそこは武器庫だった")
    back_room = Room("奥の部屋", "そこには人型のナニカがいた。顔の全面を覆う禍々しい仮面をつけたソレは人の形をしているが決して人間では無いと本能が告げている。。突然ソレはこちらに切りかかってきた。")
    back_room.event = "ソレが探索者に語り掛けてくる。─────貴様には大切な者の為に修羅に堕ちる覚悟はあるか──そう言いながらソレはこちらに切りかかってくる。"
    # 和室での初期のロール
    while True:
        print("あなたはここで『聞き耳』で振ってください。")
        print("『聞き耳』ロールを振りますか？")
        choice = input("'はい': ")

        if choice == 'はい':
            roll_result = roll_dice(1, 100)
            print(f"ロール結果: {roll_result} (聞き耳: {player.kikimimi})")
            if roll_result < player.kikimimi:
                print("【成功】「貴様には大切な者の為に修羅に堕ちる覚悟はあるか」という声が聞こえた。")
            else:
                print("【失敗】「貴様には───の為───に堕───覚悟はあるか」という声が聞こえた。突如聞こえた不気味な声にSAN値チェック 0or1d3")
                roll_result = roll_dice(1, 100)
                print(f"ロール結果: {roll_result} (SAN値: {player.sanity})")

                if roll_result < player.sanity:
                    print("【成功】SAN値減少なし。")
                else:
                    roll = roll_dice(1, 3)
                    print(f"【失敗】SAN値-{roll}。")
                    player.sanity -= roll
                    print(f"あなたのSAN値は{player.sanity}")
            break
        else:
            print("無効な入力です。もう一度試してください。")

    print(" ")
    print("そんな声が聞こえた瞬間、貴方の視界は白黒に明滅し目眩と共に意識が途切た。")
    print(" ")
    print("貴方は目覚めた瞬間見知らぬ和室に倒れていた。")
    print("服はいつも通りの普段着のままだが身につけた装飾品や手荷物は何も無い。")
    print("何はともあれあなたはここから脱出するために探索を始めた。")
    print(" ")
    # 初期の部屋を設定
    current_room = main_room

    while True:
        current_room.explore()
        current_room.actions()

        action_choice = input("アクションの番号を入力してください（1-3）: ")

        if action_choice == '1':
            if not current_room.item_searched:
                print(f"{current_room.name}を調べています...")
                # 各部屋での調査結果
                if current_room.name == "和室":
                    print("貴方は机の裏に貼り付けてあった血文字で書かれた日記の切れ端を見つける。。")
                    print(" ")
                    print(" ")
                    print(" ")
                    print("憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い")
                    print("憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い憎い")
                    print("ニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニク")
                    print("ニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニク")
                    print("ニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニクイニク")
                    print("許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな")
                    print("許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな許すな")
                    print("ユルスナユルスナユルスナユルスナユルスナユルスナユルスナユルユルスナユルスナ")
                    print("ユルスナユルスナユルスナユルスナユルスナユルスナユルスナユルユルスナユルスナ")
                    print("ユルスナユルスナユルスナユルスナユルスナユルスナユルスナユルユルスナユルスナ")
                    print(" ")
                    print(" ")
                    print("そこには知らない誰かの呪言が書かれていた。SANチェック1d3or1d8。 ")
                    roll_result = roll_dice(1, 100)
                    print(f"ロール結果: {roll_result} (SAN値: {player.sanity})")

                    if roll_result < player.sanity: # fix: Indentation corrected
                        print(f"【成功】SAN値-{roll_dice(1,3)}。")
                    else:
                        roll = roll_dice(1, 8)
                        print(f"【失敗】SAN値-{roll}。")
                        player.sanity -= roll
                        print(f"あなたのSAN値は{player.sanity}")
                        print(" ")
                        print(" ")

                elif current_room.name == "右の部屋":
                    print("いくつかの本棚がある。")
                    print("知識成功で何か見つけられるかもしれない。")
                    roll_result = roll_dice(1, 100)
                    print(f"ロール結果: {roll_result} (知識: {player.knowledge})")

                    if roll_result < player.knowledge: # fix: Indentation corrected
                        print(" ")
                        print("貴方は古ぼけた日記をみつけますが大部分が損傷しておりほとんど読めません。")
                        print(" ")
                        print("17××年4月8日")
                        print("今日から殿に仕えることになった先代の借りを返す為にも忠義を尽くさねば。")
                        print("")
                        print("17××年5月10日")
                        print("今日はいよいよ合戦だ手柄を求めてさあ行くぞ。")
                        print("")
                        print("17××年6月2日")
                        print("合戦が終わった我々の勝利だ。")
                        print("")
                        print("17××年10月24")
                        print("妻ができた私にはもったいないできた妻だ。")
                        print("")
                        print("17××年5月30日")
                        print("子供が生まれた女の子だったこれ以上に嬉しいことはない")
                        print("")
                        print("17××年7月22日")
                        print("最近妙な夢を見る、誰かが「貴様には───の為───に堕───覚悟はあるか」と問いかけてくる夢だ。")
                        print("")
                        print("17××年8月12日")
                        print("最近治安が悪い、妻と娘が心配だ。")
                        print("")
                        print("17××年9月28日")
                        print("また戦が始まるこの頃は戦ばかりだ。")
                        print("")
                        print("17××年10月4日")
                        print("戦況が悪化しているこのままでは生きて帰れないかもしれない。")
                        print("")
                        print("17××年10月5日")
                        print("覚悟を決めた。大切な者の為に私は修羅に堕ちよう。")
                        print("")
                        print("")
                        print("17××年6月1日")
                        print("ヤット帰レル、愛スルモノのモトニ")
                        print("")
                        print("17××年6月2日")
                        print("妻と娘がシンデイタ。")
                        print("")
                        print("")
                        print("")
                        print("")
                        print("")
                        print("コレデハナンノイミモナイ")
                        print("ソレナライッソ")
                        print("")
                        print("")
                        print("スベテコワシテシマエ")
                        print("")

                    else:
                        print(" ")
                        print("特に何も見つからなかった")
                        print(" ")

                elif current_room.name == "左の部屋":
                    print("武器庫だ。")
                    print(" ")
                    print("しかし、扱いが難しそうな武器ばかりだ。")
                    print("あなたが扱えそうな武器は見つからない")
                    print(" ")

                elif current_room.name == "奥の部屋":
                    print("あなたは修羅と戦闘することに決めた。")
                    print("戦闘ロールを２回連続成功させれば倒せそうだ")
                    # 戦闘開始
                    roll_results = []  # ロール結果を格納するリスト
                    for _ in range(2):
                        print("戦闘ロールを振ります...")
                        roll_result = roll_dice(1, 100)
                        roll_results.append(roll_result)  # 結果をリストに追加
                        print(f"ロール結果: {roll_result} (戦闘: {player.sentou})")

                # 2回のロール結果を表示
                    print("\n戦闘結果:")
                    for i, result in enumerate(roll_results):
                        if result < player.sentou:
                            print(f"  {i+1}回目: 成功！")
                        else:
                            print(f"  {i+1}回目: 失敗...")

                # 結果判定
                    if all(result < player.sentou for result in roll_results):  # 全て成功？
                        print("あなたは修羅を倒した！")
                        print("いつの間にか貴方が斬り伏せたソレの死体は消失しており,そこには修羅が付けていた仮面だけが残っていた。")
                        print("あなたはこう思った。")
                        print("「あの修羅の力の源はこの仮面では無いのだろうか」")
                        if "「呪物浄化の札」" in player.inventory: # インデントを修正
                            while True:
                                choice = input("「呪物浄化の札」を使いますか？ (はい/いいえ): ")
                                if choice.lower() == "はい":
                                    print("【呪物浄化の札を使う】")
                                    print("")
                                    print("貴方は呪物浄化の札を使い仮面を浄化する。仮面は白い光を放ち呪いが解けていく……")
                                    print("次の瞬間貴方が目を覚ましたのは病院だった、友人や家族から聞いた内容によると探索者は数日間、意識不明の状態だったらしい。")
                                    print("あれは夢だったのだろうか、いやそんなことは無い、なぜならあの仮面が手元にあるのだから")
                                    print("")
                                    print("エンド1「修羅には堕ちず」")

                                    break
                                elif choice.lower() == "いいえ":
                                    print("【仮面を手に取る】")
                                    print("貴方は力を求めてまたは好奇心故かはたまた何かに魅せらてか仮面を手に取る。")
                                    print("その瞬間貴方の脳内に何かが語り掛けてくる.")
                                    print("──貴様には大切な者の為に修羅に堕ちる覚悟はあるか──")
                                    print("瞬間貴方の中に何かが流れ込んでくる。")
                                    print("あなたの意識は遠く消えていった。")
                                    print("")
                                    print("エンド4「修羅に堕ちる」")
                                    break
                                else:
                                    print("無効な入力です。もう一度試してください。")
                        else: # インデントを修正
                            print("【仮面を手に取る】")
                            print("貴方は力を求めてまたは好奇心故かはたまた何かに魅せらてか仮面を手に取る。")
                            print("その瞬間貴方の脳内に何かが語り掛けてくる.")
                            print("──貴様には大切な者の為に修羅に堕ちる覚悟はあるか──")
                            print("瞬間貴方の中に何かが流れ込んでくる。")
                            print("あなたの意識は遠く消えていった。")
                            print("")
                            print("エンド4「修羅に堕ちる」")
                            break
                    else: # インデントを修正
                        print("")
                        print("修羅の剣技があなたの体を切り裂き、ついにあなたは倒れてしまう。")
                        print("貴方の意識は完全に途切れた。")
                        print("")
                        print("あなたは死んでしまった")
                        print("エンド3「修羅の刀」")



                    break  # 戦闘終了後、ゲームループを抜ける


                current_room.item_searched = True  # アイテムを調べたフラグを立てる
            else:
                print("この部屋はすでに調査済みです。")

        elif action_choice == '2':
            # 部屋ごとのアクションを実行
            if current_room.name == "和室":
                print("本棚を調べています...")
                print("目星成功で何か見つけられるかもしれない。")
                roll_result = roll_dice(1, 100)
                print(f"ロール結果: {roll_result} (目星: {player.mebosi})")

                if roll_result < player.mebosi: # fix: Indentation corrected
                        print(" ")
                        print(f"【成功】1冊の本をみつける、本にはひとつの魔術が載ってあった。。")
                        print("その本を読んでいると1枚御札のようなものが落ちてきた。アイテム：「呪物浄化の札」を入手")
                        player.add_item("「呪物浄化の札」")
                        player.display_inventory()
                else:
                        print(" ")
                        print("めぼしいものは見つからなかった")
                        print(" ")

            elif current_room.name == "奥の部屋":
                print("逃げる：あなたは急いでこの部屋を出ようとしたが、隠し扉は固く閉ざされていた！")
                print("修羅の剣技があなたの背を切り裂いた！あなたは倒れてしまう。")
                print("あなたの意識は完全に途切れた。")
                print("")
                print("あなたは死んでしまった")
                print("エンド3「修羅の刀」")

                break



        elif action_choice == '3':
            print(f"{current_room.name}を出ます。")
            # 部屋を出て次の選択へ
            while True:
                print("どの部屋に行きますか？")
                print("1: 右の部屋")
                print("2: 左の部屋")
                print("3: 奥の部屋")
                print("4: 和室に戻る")
                room_choice = input("部屋の番号を入力してください（1-4）: ")

                if room_choice == '1':
                    current_room = right_room
                    break
                elif room_choice == '2':
                    current_room = left_room
                    break
                elif room_choice == '3':
                    current_room = back_room
                    break
                elif room_choice == '4':
                    current_room = main_room
                    break
                else:
                    print("無効な入力です。もう一度試してください。")

        else:
          print("無効な入力です。もう一度試してください。")

# ゲームを開始
if __name__ == "__main__":
    text_adventure_game()
```
